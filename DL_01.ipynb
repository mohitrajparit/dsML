##Face detection with dlib and hog


```python
import cv2 as cv
import numpy
import dlib
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv.imread('people2.jpg')

cv2_imshow(img)
```

```python
face_detector_hog=dlib.get_frontal_face_detector()
```

```python
detections=face_detector_hog(img,1)
```

```python
for faces in detections:
  l,t,r,b=faces.left(),faces.top(),faces.right(),faces.bottom()
  cv.rectangle(img,(l,t),(r,b),(0,255,0),2)
cv2_imshow(img)
```

##HOG X CNN


```python
# !pip uninstall dlib
# !sudo pip install -v --install-option="--no" --install-option="DLIB_USE_CUDA" dlib
cnn_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
```

```python
detections=cnn_detector(img,1)
detections = cnn_detector(img, 1)
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(c)
  cv.rectangle(img, (l, t), (r, b), (255, 255, 0), 2)
cv2_imshow(img)
```
