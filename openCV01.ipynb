```
#converting to graysclae

import cv2
import requests
from io import BytesIO
import numpy as np
from google.colab.patches import cv2_imshow # Import the alternative for cv2.imshow()

# Fetch the image from the URL
response = requests.get('https://www.videoproc.com/images/vp-seo/convert-image-sequence-to-video-online.jpg')

# Load the image from the response content
image = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)
cv2_imshow(image)
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using cv2_imshow()
cv2_imshow(gray_image) # Use cv2_imshow instead of cv2.imshow
# cv2.waitKey(0) # Not needed with cv2_imshow
# cv2.destroyAllWindows() # Not needed with cv2_imshow
```
